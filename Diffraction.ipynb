## 1. Introduction

<u>Objective:</u> To provide a foundational understanding of diffraction, its historical development, and its fundamental principles.

**Diffraction** is the phenomenon when waves bend around obstacles or propagating through narrow openings, changing direction and deviating from geometric optics predictions.

- It is a universal wave phenomenon and is characterized by the same laws when observing wave fields of different natures, including [light waves](#2-diffraction-of-light), sound waves, water waves, and even matter waves (like [electrons](#3-electron-diffraction)).


- Diffraction is related with the phenomenon of interference, which is formulated in *The Huygens–Fresnel principle* - the main postulate of wave theory, describing and explaining the mechanism of wave propagation.

---
```
Each point on the wave front is a source of secondary spherical waves, which, starting from different points, mutually interfere. The sum of these spherical waves forms a new wavefront.
```

The general principles of diffraction modelling based on solving the wave equation:

$$ \nabla^2 \Psi(\mathbf{R}, t) - \frac{1}{v^2} \frac{\partial^2 \Psi(\mathbf{R}, t)}{\partial t^2} = 0 \tag{1}$$

where, wave function $\Psi(\mathbf{R}, t)$ desctibe wave propagation behavior (Reflection, Refraction, Diffraction, Interference,
Attenuation). For empty space with a single point source the solution is a spherical wave:

$$ \Psi(\mathbf{R}, t) = \Psi_0 \frac{e^{i(kR - \omega t)}}{kR}  \tag{2}  $$

where $R$ is the distance between observation point and source one, $k = 2\pi/\lambda$ is the wavevector and $\omega = ck$, $c$ is the light velocity, $\lambda$ is the wavelength.


In a stationary case, finding the amplitude of the diffracted wave on the aperture at the observation point accoding the Huygens-Fresnel principle, leads to integration over the aperture area, $S_\text{aperture}$, through which the wave propagates:
$$ \Psi(\mathbf{R}) = \frac{i}{\lambda} \iint\limits_{S_\text{aperture}} \frac{e^{ik|\mathbf{R} - \mathbf{R'}|}}{|\mathbf{R} - \mathbf{R'}|} \, d^2 R' \tag{3}$$

where $\mathbf{R'}$ is the coordinate of a point in the aperture, $\mathbf{R}$ is the observation point.

This allows us to calculate the amplitude and phase of the wave at any point, based on the state of the wave front at the initial stage and take into account the contribution of each elementary section to the creation of the wave field at a given point.


Today, various analytical models and numerical methods are used to calculate the diffracted field based on this approach. The complexity of solving the problem depends on many factors, such as aperture shape, boundary conditions, desired accuracy etc., so solutions are often achievable only under specific conditions.

Under these specified conditions, we can derive commonly used approximate solutions for the diffraction field:

- **Fraunhofer** approximation, which is applicable to the far field and for plane waves. This approximation is suitable for far-field diffraction, where the distances involved are large compared to the size of the aperture. 

- **Fresnel** diffraction approximation, which is applicable to the near field and for spherical waves. This approximation is useful for analyzing diffraction patterns close to the aperture.

- **Rayleigh-Sommerfeld** provides a more general approach to diffraction, incorporating the exact propagation of waves for both near-field and far-field diffraction problems using integral equations to describe the diffraction pattern. While this method can be more computationally intensive, it offers detailed results.

- **Numerical Methods** such as finite-difference time-domain (FDTD) methods, boundary element methods (BEM), or other numerical techniques, can model diffraction patterns. These methods are suitable for complex geometries and configurations where analytical solutions are difficult or impossible.

The choice of approach depends on the system parameters, such as the wavelength, aperture size, and distance from the aperture to the observation point ([see 2.1](#21-light-diffraction-based-on-single-slit-diffraction)).

###  2.1. Light Diffraction Based on Single-Slit Diffraction

Diffraction of light through a single slit is one of the fundamental phenomena in optics and wave physics. This process demonstrates the interaction of light with an obstacle (a slit). As a result of diffraction through a single slit, light behaves like a wave, interfering with itself and creating a complex pattern of maxima and minima in light intensity (Fig.2).

---

---

When a wavefront passes through a single slit of width $a$, each point within the slit can be considered as a source of secondary waves, according to Huygens' principle. These waves interfere with each other, and the resulting diffraction pattern depends on the differences in path lengths that the waves pass to reach a point on the screen.

If we divide the open part of the wave surface in the region of the slit into Fresnel zones. The width of each zone is chosen so that the difference in wave paths from the edges of these zones is equal $\lambda/2$.

The amplitude of the resulting oscillation at the observation point on the screen will be determined by the number of Fresnel zones that fit across the slit width and depends on the diffraction angle $\varphi$.

Thus, a **diffraction minimum** is observed if the number of zones is **even**:

$$ a\sin\varphi = \pm m \lambda \tag{4}$$

in this case, the amplitude of the resulting oscillation at the point is equal to zero since the oscillations from a pair of neighboring zones cancel each other out.

If the number of Fresnel zones is **odd** - a **diffraction maximum** is observed:

$$ a\sin\varphi = \pm (2m + 1)\frac{\lambda}{2}, \tag{5}$$

where $m = 1,2,3...$ order of diffraction. 

In the case where $\varphi = 0$, all secondary waves follow identical optical paths from the slit to the observation point. Consequently, a maximum with the greatest intensity will be observed at this point.

###  2.2. Light Diffraction Modeling

<u>Objective:</u> is to demonstrate approaches and methods for calculating light diffraction through a single slit.

When calculating the **diffraction intensity** distribution, the choice of method depends on various factors such as the characteristics of the obstacle, the wavelength of the incident light, the distance between the obstacle and the observation point. These factors significantly influence the complexity of the diffraction pattern and determine which analytical or numerical approach is most suitable. 

Now, let's consider the methods used to solve the problem of light diffraction through a single slit, their accuracy, and the conditions under which they are applied.

#### 2.2.1. Fraunhofer diffraction approximation

**Approximation**: The screen is positioned far from the slit, allowing the light rays to be approximated as parallel. This means both the light source and the screen (where the diffraction pattern is observed) are effectively at infinite distances from the aperture, ensuring that the phase of the light is uniform across all points on the aperture.

$$ L >> \frac{a^2}{\lambda} $$

Consider a point on the screen at an angle $\varphi$. The waves originating from different points of the slit will have different path lengths when they arrive at this point. 

$$ |r - r'| = a \sin(\varphi)$$

To calculate the total electric field $E(\varphi)$ on the screen at an angle $\varphi$, we sum the contributions from all points along the slit. This is achieved by integrating across the slit width $ a $, with the position $ y $ ranging from $ -a/2 $ to $ +a/2 $.



$$E(\varphi) = A_0 \int_{-\frac{a}{2}}^{\frac{a}{2}} e^{i \frac{2\pi}{\lambda} y \sin(\varphi)} \, dy, $$

where,

$A_0$ - the amplitude of the light passing through the slit at any point;
      
$\frac{2\pi}{\lambda} y \sin(\varphi)$ - a phase difference between the waves originating from two points separated by a small distance $dy$ within the slit.

The integration gives the amplitude of the electric field at an angle $\varphi$: 

$$E(\varphi) = A_0\frac{sin( \frac{\pi a}{\lambda} sin(\varphi))}{ \frac{\pi a}{\lambda} sin(\varphi)}$$ 

so the intensity of the light is:

$$ I(\varphi) = I_0\left(\frac{sin(\beta)}{\beta}\right)^2 \tag{6}$$

where:

$ \beta  = \frac{\pi a}{\lambda} sin(\varphi)$;  
$I_0$ is the intensity of the incident wave;  
$\varphi$ is the angle between the direction of observation and the normal to the diffracting aperture or obstacle, that depend on the width of the slit $a$ and the wavelength of the wave $\lambda$.



#### 2.2.1. Solving the Diffraction Integral with the Fast Fourier Transform (FFT) 


Consider a system consisting of a sheet with a slit (or multiple slits) and a screen parallel to the sheet, located at a distance $L$ (Fig.2). In this setup, we can position the sheet in the xy-plane, and the distance between a point on the slit and the screen is given by:


$$ |\mathbf{R} - \mathbf{R'}| = \sqrt{(x'-x)^2+(y'-y)^2+L^2}, \tag{7}$$

where $x'$ and $y'$ are the two-dimensional coordinates of the slit ($z' = 0$), $x$ and $y$ are the two-dimensional coordinates of the screen ($z = L$).

Thus, the integral in equation (3) can be approximately written as a sum:

$$ \Psi(x,y) \propto \sum_i\sum_j \frac{e^{ik \sqrt{(x'_i-x)^2+(y'_j-y)^2+L^2} }}{\sqrt{(x'_i-x)^2+(y'_j-y)^2+L ^2}}, \tag{8}$$

where $i$ and $j$ are the indices of the points forming the slit.

For the case $L^2 ≫ (x'-x)^2+(y'-y)^2$, equation (7) can be simplified to:

$$ |\mathbf{R} - \mathbf{R'}| \approx L + \frac{(x'-x)^2+(y'-y)^2}{2 L}, \tag{9}$$
Thus Eq.(3) takes the form:

$$ \Psi(x,y) \propto \iint\limits_{S_\text{aperture}} \frac{e^{ik \left(L + \frac{(x'-x)^2+(y'-y)^2}{2 L}\right)}}{L} d^2R' = \frac{e^{ik L}}{L} \iint\limits_{S_\text{aperture}} e^{i\frac{k}{2 L}((x'-x)^2 + (y'-y)^2)} d^2R', \tag{10}$$
where $ |\mathbf{R} - \mathbf{R'}| \approx L$ is used for the denominator in Eq.(3).

The equation (10)  can be solved using Fourier transformation (FT):

$$ \Psi(x,y) \propto \iint\limits_{S_\text{aperture}} e^{i\frac{k}{2 L}(x'^2-2x'x+x^2 + y'^2-2y'y+y^2)} d^2R' = e^{i\frac{k}{2 L}(x^2 + y^2)} FT_{2D}\left( f(x', y') e^{i\frac{k}{2 L}(x'^2 + y'^2)}\right), \tag{11}$$

where $f(x, y) = 1$ for $x, y \in S_{\text{aperture}}$ and 0 otherwise. The $FT_{2D}$ reciprocal space (2D FT) with coordinates $\kappa_x$ and $\kappa_y$ is associated with $x k / L$ and $y k / L$, respectively, so that the coordinates of points on the screen are $x = \kappa_y L /k$ and $y = \kappa_y L /k$. $\kappa_x \in [-\pi/L_x, \pi/L_x]$ and $\kappa_y \in [-\pi/L_y, \pi/L_y]$ for a region of size $Lx, Ly$ on the sheet including the slit (or slits).


````python
import numpy as np
from scipy.fftpack import fft2, fftshift
import matplotlib.pyplot as plt

nm = 1e-9
um = 1e-6
mm = 1e-3
cm = 1e-2
m = 1
````

````python
class Sheet():
    def __init__(self,extent_x, extent_y, Nx, Ny):
        self.Nx = int(Nx)
        self.Ny = int(Ny)

        self.x = np.linspace(-extent_x/2, extent_x/2, self.Nx)
        self.y = np.linspace(-extent_y/2, extent_y/2, self.Ny)
        self.xx, self.yy = np.meshgrid(self.x, self.y)


        self.mask = np.zeros((self.Ny, self.Nx))

    def rectangular_slit(self,x0, y0, lx, ly):
        """
            Creates a slit centered at the point (x0, y0) with width lx and height ly.

            Parameters:
            - x0, y0: Center coordinates of the slit.
            - lx: Width of the slit in the x-direction.
            - ly: Height of the slit in the y-direction.
        """
        # Apply the slit to the E field using a mask
        mask = ((self.xx > (x0 - lx/2)) & (self.xx < (x0 + lx/2)) &
                (self.yy > (y0 - ly/2)) & (self.yy < (y0 + ly/2)))

        self.mask += np.where(mask, 1, 0)

    def show_diffraction(self, Intensity2D, Intensity0_x, x_screen_display, y_screen_display, x_screen, y_screen, z):
        """
            Plotting a figure consisting of three subplots:
                ax1 - the shape of the slit;
                ax2 - the 2D distribution of the intensity of the diffraction pattern in the XY plane;
                ax3 - the intensity of light on the screen as a function of distance from the central maximum (when φ=0) 
            
            Parameters:
            - Intensity2D: 
            - Intensity0_x: 
            - x_screen_display:
            - y_screen_display:
            - x_screen:
            - y_screen:
            - z - 
         
        """
        #plot with matplotlib
        fig = plt.figure(figsize=(5, 7))

        ax1 = fig.add_subplot(3,1,1)
        ax3 = fig.add_subplot(3,1,3)
        ax2 = fig.add_subplot(3,1,2,sharex=ax3)

        # display figure
        ax1.imshow(self.mask, aspect="equal", extent=[min(self.x), max(self.x), min(self.y), max(self.y)])
        ax1.set_xlabel("x (m)")
        ax1.set_ylabel("y (m)")

        ax2.imshow(Intensity2D, aspect="equal", extent = [x_screen[0], x_screen[-1], y_screen[0], y_screen[-1]], cmap ='gray')
        ax2.set_ylabel("y (m)")
        ax2.set_ylim([-y_screen_display / 2, y_screen_display / 2])
        ax2.set_xlim([-x_screen_display / 2, x_screen_display / 2])

        # I(x, y = 0)
        Intensity_x = Intensity2D[self.Ny//2]
        ax3.plot(x_screen, Intensity_x, linewidth = 1, label="Intensity")

        #calculation using analytical formula (subplot ax3, red line)
        if Intensity0_x is not None:
            ax3.plot(x_screen, Intensity0_x, ":", label="Intensity_0")
            ax3.legend()

        ax3.set_xlabel("x (m)")
        ax3.set_ylabel("Intensity (a.u.)", labelpad=20)
        # ax3.set_yticklabels([])
        ax3.set_xlim([-x_screen_display / 2, x_screen_display / 2])

        # Hide x-axis tick labels for ax2
        plt.setp(ax2.get_xticklabels(), visible=False)

        # Adjust figure to correct the figure
        data_aspect_ratio1 = (max(self.y) - min(self.y)) / (max(self.x) - min(self.x))
        data_aspect_ratio2 = (z/2) / (z)

        pos3 = ax3.get_position()  # Get the original position of ax3

        fig_width, fig_height = fig.get_size_inches()  # Figure dimensions in pixels
        fig.set_size_inches((fig_width, int((data_aspect_ratio1 + data_aspect_ratio2 + 0.25) * fig_width)))

        y3 = pos3.y0
        ax3.set_position([pos3.x0, y3, pos3.width, pos3.width *0.25])
        y3 += 0.01 + pos3.width * 0.25
        ax2.set_position([pos3.x0, y3, pos3.width, pos3.width * data_aspect_ratio2])
        y3 += 0.01 + pos3.width * data_aspect_ratio2 + pos3.y0
        ax1.set_position([pos3.x0, y3, pos3.width, pos3.width * data_aspect_ratio1])

        plt.show()
````

````python
#calculation using analytical formula (eq.6)
def diffraction_Franhover(k, slit_width, slit_height, x_screen, y_screen, z):
    X, Y =  np.meshgrid(x_screen, y_screen)
    beta_x = (k * slit_width / 2) * X/z + 1e-10
    beta_y = (k * slit_height / 2) * Y/z + 1e-10
    return (np.sin(beta_x) / beta_x) ** 2 * (np.sin(beta_y) / beta_y) ** 2

#numerical calculation: using Fast Fourier Transform (eq.11)
def diffraction_FFT(k, sheet, z):
    intensity = fftshift(fft2(sheet.mask * np.exp(1j * k/(2*z) * (sheet.xx**2 + sheet.yy**2))))
    intensity = np.absolute(intensity)**2
    return intensity / intensity.max()

#direct integtation (eq.8)
def diffraction_direct(k, sheet, x_screen, y_screen, z):
    N_x, N_y = x_screen.size, y_screen.size

    # Initialize the field on the screen
    field = np.zeros((N_y, N_x), dtype=np.complex128)

    # Define the integrand function
    def integrand(k, slit_x, slit_y, x, y, z):
        """
        integrand function consisting of spherical wave
        """
        # Calculate the distance from the slit point to the screen point
        kr = k * np.sqrt((x - slit_x)**2 + (y - slit_y)**2 + z**2)
        return np.exp(1j * kr) / kr

    # Loop over each point in the slit and calculate its contribution to the screen
    for ix in range(N_x):
        for iy in range(N_y):
            # Calculate the contribution to the field at the screen point
            field[iy][ix] = integrand(k, sheet.xx, sheet.yy, x_screen[ix], y_screen[iy], z).sum()

    # Calculate the intensity on the screen (square of the field magnitude)
    intensity = np.abs(field)**2

    # Normalize the intensity
    return intensity / intensity.max()
````

````python
"""
    Simulation input
"""
# calculation area (sheet) around slit in (m)
Lx = 10*2000 *4 * nm
Ly = 10*600*4 * nm
# number of point on screen (for FFT it is better to use odd sampling)
Nx = 151
Ny = 81

# distance from slit to the screen (m)
z = 0.001 * m
# observation area in screen
screen_width = z
screen_height = z

# distance between the slits, if the calculation for two slits
# D = 128 * nm

# wavelength (m)
λ = 700 * nm
k = 2 * np.pi / λ

sheet = Sheet(extent_x=Lx , extent_y=Ly, Nx=Nx, Ny=Ny)

# Create a rectangular slit at the center
slit_width = 2000*4 * nm
slit_height = 300*4 * nm
sheet.rectangular_slit(x0=0, y0=0, lx=slit_width, ly=slit_height)

# if the calculation for two slits you need to add another one
# by changing the coordinates accordingly, for example:

# sheet.rectangular_slit(x0 = -D/2, y0 = 0, lx = 22 * mm , ly = 88 * mm)
# sheet.rectangular_slit(x0 = +D/2, y0 = 0, lx = 22 * mm , ly = 88 * mm)

#screen size mm
dx_screen = z*λ/Lx
dy_screen = z*λ/Ly
x_screen = np.linspace(-dx_screen * ((Nx if Nx % 2 else Nx + 1) // 2), dx_screen * ((Nx if Nx % 2 else Nx - 1) // 2), Nx)
y_screen = np.linspace(-dy_screen * ((Ny if Ny % 2 else Ny + 1) // 2), dy_screen * ((Ny if Ny % 2 else Ny - 1) // 2), Ny)

#calculation using analytical formula (subplot ax3, red line)
intensity0 = diffraction_Franhover(k, slit_width, slit_height, x_screen, y_screen, z)

#numerical calculation using Fast Fourier Transform (subplot ax3, black line)
intensity_FFT = diffraction_FFT(k, sheet, z)


# slit integration methods
slit = Sheet(extent_x=slit_width , extent_y=slit_height, Nx=Nx, Ny=Ny)


# Define the slit and screen coordinates
x_screen = np.linspace(-screen_width / 2, screen_width / 2, Nx)
y_screen = np.linspace(-screen_height / 2, screen_height / 2, Ny)

intensity0_2 = diffraction_Franhover(k, slit_width, slit_height, x_screen, y_screen, z)
# numerical calculation using direct integration
intensity_direct = diffraction_direct(k, slit, x_screen, y_screen, z)
````

````python
# display intencities
sheet.show_diffraction(intensity_FFT, intensity0[sheet.Ny//2], screen_width, screen_height, x_screen, y_screen, z)
sheet.show_diffraction(intensity0, None, screen_width, screen_height, x_screen, y_screen, z)
slit.show_diffraction(intensity_direct, intensity0_2[sheet.Ny//2], screen_width, screen_height, x_screen, y_screen, z)
````

The Fraunhofer model assumes that the particles are spherical, opaque, and have a diameter much larger than the wavelength of the laser beam (Fig.4(b)), while the Mie theory considers spherical, transparent particles that absorb part of the light, with a size comparable to the laser's wavelength (Fig.4(c)).

Let's consider a case where the Fraunhofer approximation can be applied (when the particle size is much larger than the laser wavelength), where only scattering at the particle's contour in the near-forward direction is taken into account. Assuming that the diffraction pattern of the particle is equivalent to that of a thin, two-dimensional circular disk, derive an expression for the intensity as a function of the diffraction angle:

$$ I(\theta) = \frac{I_0}{k^2a}\alpha^4\left[\frac{J_1(\alpha\sin\theta)}{\alpha\sin\theta}\right]^2, \tag{12} $$
where $J_1$ is the Bessel function of the first kind of order unity ($x_1 \approx 3.8317$) and $\alpha = \pi d/\lambda$ - dimensionless size parametr.

From equation (12), we can determine the position of the first minimum in the diffraction pattern formed when light passes through a circular aperture (the Airy disk pattern), and thus find the diameter of the particles $d$, see Fig.5:

$$ \sin\theta = \frac{1.22 \lambda} {d}$$

### 3.1. The Fundamentals of Electron Diffraction

**Electron diffraction** is a general term for phenomena involving changes in the direction of electron beams due to elastic interactions with atoms [John M., Cowley (1995). Diffraction physics. Elsevier. ISBN 0-444-82218-6. OCLC 247191522]:


- When a beam of electrons is directed at a material, it interacts with the material's atoms through elastic scattering (meaning no energy is lost by the electrons).

- The atomic planes within a crystal act as a diffraction grating for the electrons , similar to how optical diffraction occurs with light passing through a slit or grating. ### 3.2. Electron Diffraction Modelling

<u>Objective:</u> To develop a deeper understanding of the underlying physics of electron diffraction by simulating electron diffraction patterns and analyzing the effects of various parameters on the temporal evolution of electrons in space.

The temporal evolution of electrons in space is described of time-dependent Schrodinger equation:
$$
i\hbar {\frac {d}{dt}} \Psi ={\hat {H}} \Psi, \tag{14}
$$
The solution of Eq.(14) can be written in operator form:
$$
\Psi(t) = e^{ -\frac{i}{\hbar} {\hat {H}} t} \Psi(t=0), \tag{15}
$$
where $\Psi(t)$ is the time evolution of the wave function, starting from the initial state $\Psi(t=0)$.

The exponential operator in Eq.(15) can be expanded in terms of Chebyshev polynomials. The Chebyshev expansion method is often used to approximate functions of operators, especially for large-scale problems, due to its numerical efficiency and stability. The core idea is to expand the exponential operator as a series of Chebyshev polynomials $T_n(x)$, where $x$ is a rescaled version of the Hamiltonian $\hat{H} / \hbar$, ensuring that the spectrum of $x$ lies in the interval $[-1, 1]$, which is the domain of Chebyshev polynomials.

The function $f(x) = e^{-i x t}$ can be expanded in terms of Chebyshev polynomials $T_n(x)$:
$$
   f(x) = e^{-i x t} \approx \sum_{n=0}^{N} c_n T_n(x),
$$

where $T_n(x)$ are Chebyshev polynomials of the first kind, and the coefficients $c_n$ are given by:
$$
   c_n = \frac{2}{\pi} \int_{-1}^{1} \frac{e^{-i t x} T_n(x)}{\sqrt{1 - x^2}} dx = i^n J_n(t),
$$
where $J_n(t)$ are the Bessel functions of the first kind.

Once the expansion coefficients $c_n$ are determined, the time-evolved state can be computed as a linear combination of Chebyshev polynomials applied to the rescaled Hamiltonian $\hat{H}$:
$$
   \Psi(t) \approx \sum_{n=0}^{N} c_n T_n(\hat{H}) \Psi(0).
$$
   
The action of the Chebyshev polynomials $T_n(\hat{H})$ on the initial state $\Psi(0)$ is computed recursively using the recurrence relation for Chebyshev polynomials:
   
$$
   T_0(x) = 1, \quad T_1(x) = x, \quad T_{n+1}(x) = 2x T_n(x) - T_{n-1}(x).
$$
   
So, starting with $\Psi_0 = \Psi(t=0)$ and $\Psi_1 = \hat{H} \Psi_0$, we can compute $\Psi_n = T_n(\hat{H}) \Psi_0$ recursively for $n = 2, 3, \dots, N$.

In practice, the series is truncated at a finite number $N$, which is determined by the desired accuracy and the properties of the Hamiltonian and time evolution. The error decreases exponentially with the number of terms due to the rapid convergence of the Chebyshev expansion for smooth functions.

````python
def Get_Propagator_Chebyshev_Expansion(Psi,H,a,z):
	"""
		This fucntion advances Psi in time units of delta t by 
		the Chebyshev expansion of the propagator.
	
	"""
	Hz = H*z
	g0,g1 = Psi, np.dot(Hz,Psi)
	Psi = a[0]*g0 + a[1]*g1

	for jt in range(2,len(a)):
		g0,g1 = g1, g0 + np.dot(Hz,g1)*2 # reccurence of Chebyshev polynomials
		Psi += a[jt]*g1 # accumulation step

	Psi1 = np.dot(H,Psi)
	return Psi,Psi1

def Get_Chebyshev_Coefficients(Nc,Nc_max,amin,r):
	a=np.zeros(Nc_max,float)
	C=2
	a[0] = special.jv(0,r)
	k = np.arange(1,Nc)
	a[k] = C*special.jv(k,r)

	while (a[Nc-1] > amin):
		a[Nc] = C*special.jv(Nc,r)
		Nc+=1

		if(Nc>Nc_max):
			print('The argument of the Chebyshev exp. is too large')
			break
	return a[:Nc],Nc


def Get_Initial_Wave_Packet2D(x,y,k0x,a0x,k0y,a0y):
	Ax = (2/(np.pi*a0x**2))**(1/4)*np.exp(-(x/a0x)**2)*np.exp(ci*k0x*x)
	Ay = (2/(np.pi*a0y**2))**(1/4)*np.exp(-(y/a0y)**2)*np.exp(ci*k0y*y)
	Psi=	np.einsum( 'x,y->xy',Ax,Ay )
	return Psi

# FFT for single eigen state ( psi(x,y) )
def Get_Fourrier_Transform_WF2Dsingle(psi,x,delta_x,y,delta_y,kx,ky):
	KX,X = np.meshgrid(kx,x, sparse=False, indexing='ij')
	KY,Y = np.meshgrid(ky,y, sparse=False, indexing='ij')

	psi_FT = ( (np.exp(-ci*(KX*X)) * delta_x) @ psi @  (np.exp(-ci*(KY*Y))*delta_y).T )/(2*np.pi) # Integrate Trapeze
	return psi_FT


def Get_Kinetic_Operator(x,mu,LBOXx,Type):
	""" 
		Calculation of the kinetic operator

		inputs:
			x -	vector of  coordinate
			mu -mass
			LBOXx- the length of simulation box
			Type - type of the discretization for the kinetic operator

		Outputs:
			T - 2D array of the kinetic operator 
	
	"""
def Get_Kinetic_Operator(x,mu,LBOXx,Type):
	""" 
		Calculation of the kinetic operator

		inputs:
			x -	vector of  coordinate
			mu -mass
			LBOXx- the length of simulation box
			Type - type of the discretization for the kinetic operator

		Outputs:
			T - 2D array of the kinetic operator 
	
	"""
	N = len(x)
	if(Type==0):
		coef= (np.pi/LBOXx)**2/mu
		Di	= np.arange(1,N)
		if(N%2==0):
			T_diag		= (N*N+2)/6.
			T_nondiag	= (-1.)**Di/(np.sin(Di*np.pi/N))**2
			#print('<br>even T=',coef*hstack( (T_diag, T_nondiag) ))
		else:
			T_diag		= (N*N-1)/6.
			T_nondiag	= (-1.)**Di*np.cos(Di*np.pi/N)/(np.sin(Di*np.pi/N))**2
		T = toeplitz( coef*np.hstack( (T_diag, T_nondiag) ) )
	elif(Type==1):
		# np.sin basis set
		coef	= (np.pi/LBOXx)**2/(2*mu)/2
		T = np.zeros((N,N),float)
		I1, I2 = np.meshgrid(np.arange(N),np.arange(N))
		cind = np.logical_not(I1==I2)
		DI = I1[cind]-I2[cind]
		SI = I1[cind]+I2[cind]
		T[cind] = coef*(-1.)**DI*(1./np.sin(np.pi*DI/(2*(N+1)))**2 - 1./np.sin(np.pi*(SI+2)/(2*(N+1)))**2)
		cind = (I1==I2)
		T[cind] = coef*((2*(N+1)**2+1)/3. - 1/np.sin(np.pi*(I1[cind]+1)/(N+1))**2)
	return T
	# N = len(x)
	# # sin basis set
	# coef	= (np.pi/LBOXx)**2/(2*mu)/2
	# T = np.zeros((N,N),float)
	# I1, I2 = np.meshgrid(np.arange(N),np.arange(N))
	# cind = np.logical_not(I1==I2)
	# DI = I1[cind]-I2[cind]
	# SI = I1[cind]+I2[cind]
	# T[cind] = coef*(-1.)**DI*(1./np.sin(np.pi*DI/(2*(N+1)))**2 - 1./np.sin(np.pi*(SI+2)/(2*(N+1)))**2)
	# cind = (I1==I2)
	# T[cind] = coef*((2*(N+1)**2+1)/3. - 1/np.sin(np.pi*(I1[cind]+1)/(N+1))**2)
	# return T


def Get_Kinetic_Operator2D(x,y,mu,LBOXx,LBOXy,Type=0):
	return Get_Kinetic_Operator(x,mu,LBOXx,Type), Get_Kinetic_Operator(y,mu,LBOXy,Type)


def Get_Hamiltonian_Operator2D(Nx,Ny,Tx,Ty,V):

	return   (np.einsum('ij,kl->ikjl', Tx, np.eye(Ny,Ny))+\
			np.einsum('ij,kl->ljki', Ty, np.eye(Nx,Nx))).reshape((Nx*Ny,Nx*Ny))+\
			np.diag(V.flatten()) # H[xy,xy]


def Get_E_min_max2D(V,delta_x,delta_y,mu):
	pxmax = np.pi/delta_x.min()
	pymax = np.pi/delta_y.min()
	Emax = (pxmax**2+pymax**2)/(2.*mu)+V.max()
	Emin = V.min()
	Eshift = (Emax+Emin)/2

	return Emax,Emin,pxmax,pymax,Eshift

def Get_Constant_K(Nk,delta_x):
	kmax = np.pi/delta_x.max()
	k = np.linspace(-kmax,kmax+kmax/Nk, num=2*Nk+1, endpoint=False)
	dx = np.diff(k)
	delta_k = (np.concatenate([[0],dx]) + np.concatenate([dx,[0]]))/2										
	return k,delta_k,kmax

def Get_Constant_K2D(Nkx,delta_x,Nky,delta_y):
	kx,delta_kx,kxmax = Get_Constant_K(Nkx,delta_x)
	ky,delta_ky,kymax = Get_Constant_K(Nky,delta_y)

	return kx,delta_kx,kxmax,ky,delta_ky,kymax
````

````python
# These functions used by the function Get_Profile for generation of the spatial grid.
##########################################################################################################

def Get_Grid(N,x0,xL=0,delta_x=None,Type=0):
	""" 
		These functions used by the function Get_Profile for generation of the spatial grid.
	"""
	if(Type==0):
		x=np.zeros(N,float)   # declare of the grid point
		if delta_x==None:
			LBOX = xL-x0
			x = np.linspace(x0,xL, num=N+1)
			delta_x0 = np.diff(x)
		else:
			delta_x0 = delta_x
			x = x0+np.cumsum(delta_x[:-1])
			LBOX = x.max()-x0
		x = x[:-1]
		delta_x = np.concatenate([[delta_x0[0] + delta_x0[-1]],delta_x0[:-1] + delta_x0[1:]])/2

	if(Type==1):
		x=np.zeros(N,float)   # declare of the grid point
		if delta_x==None:
			LBOX = xL-x0
			x = np.linspace(x0,xL, num=N+2)
			delta_x0 = np.diff(x)
		else:
			delta_x0 = delta_x
			x = x0+np.cumsum(delta_x[:-1])
			LBOX = x.max()-x0
		x = x[1:-1] #x0 and xL are excluded from the grid
		delta_x	= (delta_x0[:-1] + delta_x0[1:])/2

	return x,delta_x0,delta_x,LBOX


def Get_Constant_Grid(N,x0,xL,Type=0):
	""" 
		Used by the function Get_Profile for generation of the spatial grid.

	"""
	# build the grid point
	LBOX = xL-x0
	if(Type==0):
		x			= np.linspace(x0,xL, num=N+1)
		delta_x0	= np.diff(x)
		delta_x		= np.concatenate([[delta_x0[0]+delta_x0[-1]],delta_x0[:-1]+delta_x0[1:]])/2
		x = x[:-1]

	elif(Type==1):
		x			= np.linspace(x0,xL, num=N+2)
		delta_x0	= np.diff(x)
		delta_x		= (delta_x0[:-1]+delta_x0[1:])/2
		x = x[1:-1] #x0 and xL are excluded from the grid

	else:
		print("<br>Uknown input method (use Type=0 for .. and Type=1 for..), Type=",Type)

	return x,delta_x0,delta_x,LBOX


def Get_Constant_Grid2D(Nx,x0,xL,Ny,y0,yL,Type=0):
	""" 
		Used by the function Get_Profile for generation of the spatial grid for 1D and 2D cases.
	"""
	x,delta_x0,delta_x,LBOXx = Get_Constant_Grid(Nx,x0,xL,Type)
	y,delta_y0,delta_y,LBOXy = Get_Constant_Grid(Ny,y0,yL,Type)

	return x,delta_x0,delta_x,y,delta_y0,delta_y,LBOXx,LBOXy
````

````python
"""
	Etrer your data for electron diffraction simulation.

	* The value of Pot_Type determines the barrier profile.
	
	input (with defolt values):
		 Fields of InputData argument and their default values:
			Dimensionality - 2
			Type - Type of kinetic operator discretisation  - 0 plane waves basis set - in this case
															- 1 sine basis set
			Pot_Type - type of the profile of the potential energy for barriers:
															- 20 free particle
															- 31 Infinite well for 1 slit
															- 32 Infinite well for 2 slits
															- 33 Infinite well for n slits 
			V0 - Potential energy height - 1
			mu - mass - 1
			Nx - Number of points along x-axis - 100
			Ny - Number of points along y-axis - 100
			x0 - Starting x-coordinate - 0
			y0 - Starting y-coordinate - 0
			xL - Ending x-coordinate - 1
			yL - Ending y-coordinate - 1

"""

OutData	= lambda:0
InputData = lambda:0

InputData.Type = 0
InputData.Dimensionality = 2

# ******************************************************************************#
# In calculations Type = 0 is used: plane waves basis set - the method of discretization of the kinetic energy operator

# The barrier profile:	:
# 20 --> free particle
# 31 --> Infinite well for 1 slit
# 32 --> Infinite well for 2 slits
# 33 --> Infinite well for n slits does not work.. Need to fix potential with n_slits
InputData.Pot_Type = 32

# Temporal used to make different calculations:
# Temporal==0 or Temporal==1 show only barrier profile
# Temporal==2 as 0,1 + show only WP profile
# Temporal==3 run temporal evolution calculations
# Temporal==4 run transmittance calculations
InputData.Temporal	= 1

InputData.mu		= 1		# The particle mass, a.u.: 1
InputData.V0		= 100 	# The height of the quantum barrier, a.u.: 100


InputData.Nx	= 100 	# The number of spatial points for the x-axis: 100
InputData.Ny	= 50	# The number of spatial points for the y-axis: 50

InputData.x0	= -50	# The left x-position of the calculation space, a.u.: -50
InputData.xL	=  50	# The right x-position of the calculation space, a.u.: 50
InputData.y0	= -50	# The left y-position of the calculation space, a.u.: -50
InputData.yL	=  50	# The right y-position of the calculation space, a.u.: 50

InputData.slit_d	= 3	# The slit width (changes in x-axis), a.u.: 8
InputData.slit_ls	= 8		# The slit length (changes in y-axis), a.u.: 4

#If the number of slits is more than one
InputData.slit_ly 	= 8		# The distance between centres of slits (period), a.u.: 4
InputData.n_slits = 5 # The number of slits: 3


# Parameters for the initial wave packet:
#-----------------------------------------------------------------------------
	## The initial position of the wave packet along
InputData.xc_WP = -25   # the x-axis: -25
InputData.yc_WP = 0	  # the y-axis: 0

	## The wave packet width in the direct space along
InputData.a0x	= 10	# the x-axis: 10
InputData.a0y	= 30	# the y-axis: 30

	## The wave vector of the wave packet center in the inverse direct space 
InputData.k0x	= 1		# of the x-axis: 1
InputData.k0y	= 0		# of the y-axis: 0

	## The number of points for the inverse space:
InputData.Nkx	= 25  # of the x-axis: 25
InputData.Nky	= 26  # of the y-axis: 26


# Parameters for temporal evolution of the wave packet :
#-----------------------------------------------------------------------------

	## Time parameters: 
InputData.Nt = 300  		# the number of points for time discretization: 300
InputData.delta_t = 0.2 	# The time step for simulations: 0.2
InputData.Nc_max 	= 10000	# Thr maximum number of iterations: 10000


InputData.amin = 5e-7 	# The accuracy of the Chebyshev decomposition: 5e-7
InputData.epsilon	= 1e-2	# The convergence criteria: 1e-2


	##The screen position for observing the diffraction: 20
InputData.x_screen = 20

#?
InputData.omegax	= 1
InputData.omegay	= 1

# Parameters for sep-like potentials: d = 5. sigma = 1e-2
InputData.d = 5.
InputData.sigma = 1e-2

InputData.isCheckOld = True
# InputData.isCheckOld = False




#????????????????????????????????????????????
Lambda = 2 * np.pi/np.sqrt(InputData.k0x**2 + InputData.k0y**2)#max(k0x,k0y)
if not hasattr(InputData, 'ls'):
	ls = 0
	print('ls',ls)
	if(InputData.Pot_Type==31):
		ls = 2*Lambda
	elif(InputData.Pot_Type==32 or InputData.Pot_Type==33):
		ls = Lambda/2
	InputData.ls = ls
print("x-coordinate for which the diffraction spot could be observed is ls**2/Lambda=%g"%(InputData.ls**2/Lambda + (InputData.xL + InputData.x0)/2.))
print("y coordinate for which the diffraction spot could be observed is ls**2/Lambda=%g"%(InputData.ls**2/Lambda))
#????????????????????????????????????????????


## main procedure
OutData,InputData = Get_Profile(InputData)
if hasattr(OutData,'Error'): # return error message if there is an error
	print (OutData)

Emax,Emin,pxmax,pymax,OutData.Eshift = Get_E_min_max2D(OutData.V,InputData.delta_x,InputData.delta_y,InputData.mu)
OutData.V -= OutData.Eshift


if(pxmax < InputData.k0x or pymax < InputData.k0y):
	OutData.Error	= 'Attention pxgrid < k0x or pygrid < k0y : increase Nx and Ny or decrease k0x and k0y'

#		sys.exit('Attention pxgrid<k0x or pygrid<k0y : increase Nx and Ny or decrease k0x and k0y') 

#   Build the grid for FT
OutData.kx,delta_kx,_,OutData.ky,delta_ky,_	= Get_Constant_K2D(InputData.Nkx,InputData.delta_x0,InputData.Nky,InputData.delta_y0)


#   Generate an intial wavepacket
Psi			= Get_Initial_Wave_Packet2D(OutData.x-InputData.xc_WP,OutData.y-InputData.yc_WP,InputData.k0x,InputData.a0x,InputData.k0y,InputData.a0y)
Psi_FT		= Get_Fourrier_Transform_WF2Dsingle(Psi,OutData.x,InputData.delta_x,OutData.y,InputData.delta_y,OutData.kx,OutData.ky)


if InputData.Temporal < 3: # show WP and barrier profile
	OutData.psi		= abs(Psi)**2
	OutData.psi_FT	= abs(Psi_FT)**2


Tx,Ty = Get_Kinetic_Operator2D(OutData.x,OutData.y,InputData.mu,InputData.LBOXx,InputData.LBOXy,InputData.Type)

H = Get_Hamiltonian_Operator2D(InputData.Nx,InputData.Ny,Tx,Ty,OutData.V)

norm			= np.zeros(InputData.Nt, float)
H_average		= np.zeros(InputData.Nt, float)
Delta_X			= np.zeros(InputData.Nt, float)
Delta_Y			= np.zeros(InputData.Nt, float)
Delta_Px		= np.zeros(InputData.Nt, float)
Delta_Py		= np.zeros(InputData.Nt, float)
Delta_X_analytic= np.zeros(InputData.Nt, float)
Delta_Y_analytic= np.zeros(InputData.Nt, float)
x_average		= np.zeros(InputData.Nt, float)
y_average		= np.zeros(InputData.Nt, float)
px_average		= np.zeros(InputData.Nt, float)
py_average		= np.zeros(InputData.Nt, float)
Heisenberg		= np.zeros(InputData.Nt, float)
Heisenberg_X	= np.zeros(InputData.Nt, float)
Heisenberg_Y	= np.zeros(InputData.Nt, float)
time			= np.zeros(InputData.Nt, float)


#   Bessel coefficients for Chebyshev expansion
r = InputData.delta_t*(Emax - Emin)/2.
z = -ci*InputData.delta_t/r
Nc = int(max(2.0,r)) #nombre de paremtre Chebyshev de devloppement
a,Nc = Get_Chebyshev_Coefficients(Nc,InputData.Nc_max,InputData.amin,r)


OutData.Nc				= Nc
OutData.psi_t			= []
OutData.psi_FT_t		= []
OutData.psi_analytic_t	= []
time					= np.linspace(0,(InputData.Nt-1)*InputData.delta_t,InputData.Nt)

#    >>>>>>>>  the main time integration loop <<<<<<<<
for it in range(InputData.Nt):
	if it==0:
		Psi1 = np.dot(H,Psi.flatten()).reshape((InputData.Nx,InputData.Ny))
	else:
		Psi,Psi1	= Get_Propagator_Chebyshev_Expansion(Psi.flatten(),H,a,z)
		Psi			= Psi.reshape((InputData.Nx,InputData.Ny))
		Psi1		= Psi1.reshape((InputData.Nx,InputData.Ny))

	OutData.psi_t.append(abs(Psi)**2)
	OutData.psi_FT_t.append(abs(Psi_FT)**2)


	it += 1
	if it > InputData.Nt:
		it = InputData.Nt
	OutData.time				= time[:it]
	OutData.H_average			= H_average[:it]
	OutData.Delta_X_analytic	= Delta_X_analytic[:it]
	OutData.Delta_Y_analytic	= Delta_Y_analytic[:it]
	OutData.norm				= norm[:it]
	OutData.Heisenberg			= Heisenberg[:it]
	OutData.Delta_X				= Delta_X[:it]
	OutData.Delta_Y				= Delta_Y[:it]
	OutData.Delta_Px			= Delta_Px[:it]
	OutData.Delta_Py			= Delta_Py[:it]
	OutData.x_average			= x_average[:it]
	OutData.y_average			= y_average[:it]
	OutData.px_average			= px_average[:it]
	OutData.py_average			= py_average[:it]
	OutData.Heisenberg			= Heisenberg[:it]
	OutData.Heisenberg_X		= Heisenberg_X[:it]
	OutData.Heisenberg_Y		= Heisenberg_Y[:it]
````

### 3.3. Applications of Electron Diffraction

<u>Objective:</u> to demonstrate the applications of electron diffraction in studying the structures of materials using a transmission electron microscope.

**Diffraction of electrons** provides a wealth of information about the structural properties of materials, such as crystal structure and symmetry characterization, orientation and grain boundary definitions, and defect analysis. Today, electron diffraction is an integral technique of transmission electron microscope (TEM) systems and is widely employed in materials science, nanotechnology, and other fields for analyzing the structures of materials, including metals, semiconductors, and biological specimens.

In Fig. 8, a schematic illustration of the main components of a transmission electron microscope (TEM) is presented. 

The typical transmission electron microscope laboratory contains a machine with these components: <u>electron gun</u>, which generates electrons that pass down <u>the electron column</u> through <u>a electro-magnetic lens system</u> and <u>specimen/sample chamber</u> to a <u>image capture system</u>, detectors (for example, STEM) and the main control panel and operational controls. 

Maintaining a vacuum is crucial for the quality of the final image, as air particles can deflect the trajectory of electrons in the electron beam, reducing both the resolution and the signal-to-noise ratio. Transmission electron microscopy (TEM) is an analytical technique used to visualize the smallest structures in matter. Unlike optical microscopes, which rely on light in the visible spectrum, TEM can reveal the smallest detail at the atomic scale by magnifying nanometer structures up to 50 million times. This is because electrons can have a significantly shorter wavelength (about 100,000 times smaller) than that of visible light when accelerated through a strong electromagnetic field, thus increasing the microscope resolution by several orders of magnitude.



ED is a technique for examining the crystal structure from the diffraction pattern obtained by irradiating a sample with an electron beam. Crystallographic information of the substance is then obtained.

In the case of a transmission electron microscope (TEM, shown in Fig. 1), the electron diffraction pattern consists of regularly spaced spots for a single crystal, concentric rings for a polycrystal, and broad rings for an amorphous sample.

There are many types and techniques of electron diffraction.

Electron diffraction patterns give crystallographic information about a material and determine different types of materials which can be amorphous, single crystalline or polycrystalline. There are three types of electron diffraction patterns and the formation of each pattern depends on the different conditions of the specimen such as thickness, crystal structure and so on.

The polycrystalline materials exhibit ring pattern

The single crystalline materials show (a) spot pattern or (b) Kikuchi line pattern or (c) a combination of spot and Kikuchi line patterns

The spot and Kikuchi line patterns are obtained from a special area of specimen which is called the 'selected area' [5–8]. Selected area electron diffraction (SAED) is a technique in TEM to obtain diffraction patterns that result from the electron beam scattered by the sample lattice.
